```python
import numpy as np
import torch
import pandas as pd
df = pd.read_csv('mydata.csv')
df = df.values
df.astype('float32')
df = df[:,1:]
inputs1 = df[:,[0,1]]
outputs1 = df[:,[2]]
inputs1 = inputs1.astype('float32')
outputs1 = outputs1.astype('float32')
import torch.nn as nn
import torch.nn.functional as F
```

```python
class LinearRegressionModel(nn.Module):
    def __init__(self, input_dim, output_dim):
        super(LinearRegressionModel, self).__init__()
        self.linear1 = nn.Linear(input_dim, output_dim)
        #self.linear2 = nn.Linear(3, output_dim)

    def forward(self, x):
        out = self.linear1(x)
        #out = self.linear2(out)
        return out
input_dim = 2
output_dim = 1
model = LinearRegressionModel(input_dim, output_dim)
learning_rate = 0.00000001
loss_fn = nn.MSELoss()
opt = torch.optim.SGD(model.parameters(), lr=learning_rate)
```

```python
epochs =500
```

```python
for epoch in range(epochs):
    epoch += 1
    # Convert numpy array to torch Variable
    inputs = torch.from_numpy(inputs1).requires_grad_()
    outputs = torch.from_numpy(outputs1)

    # Clear gradients w.r.t. parameters
    opt.zero_grad() 

    # Forward to get output
    pred = model(inputs)

    # Calculate Loss
    loss = loss_fn(outputs, pred)
    # Getting gradients w.r.t. parameters
    loss.backward()

    # Updating parameters
    opt.step()

    print('epoch {}, loss {}'.format(epoch, loss.item()))
```

```python
torch.save(model, 'mymodel1.pt')
```
